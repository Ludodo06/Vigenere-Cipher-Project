## What is the learning activity?

For example, **"Find the Most Reliable Pitching Machine"**

The first task consists in trying to decode a text encoded using the Caesar cipher. A code encoded with the Caesar cipher is given together with a Caesar cipher wheel but no explanation is given about which tecnique has been used to encode the text and about what it is. A possibility of receiving an hint is offered; if it is accepted, the first word of the plaintext is revealed.

Later an explanation of the Caesar cipher and on how it works is done, adding what a frequency analysis is and how it can be used to easily crack a substitution cipher. 

Then a small introduction to what the Vigenère cipher is and how to use it is given, together with examples and with a  tool to encode and decode when the key is known. 

Later, an encoded text is given, without the key, and the student needs to try to decode it. Again, an hint can be given in order to help the student to achieve the task: since previously an explication of what a frequency analisis is was given, a tool that reorders the letters by dividing them in groups of the key length could be used in order to achieve the task if the concept is clear. Also knowing the size of the key could facilitate the student in the decoding of the code.

Finally once the exercise is done, it is explained how it is possible to use the Kasiski examination method to break the Vigenère cipher. 



Here are the ways in which my learning activity will activate learners' prior knowledge: 

As said previously, this problem-solving activity does not require any particular previous knowledge in order to be solved. In order to be able to decrypt the code encoded using the Caesar cipher just a knowledge of basic logic is needed. For solving the Vigenère cipher, again basic logic is needed and the concept of frequency analysis explained previously needs to be transfered to this new task.


Here are the ways in which my learning activity will create an awareness of knowledge gaps:

For solving the first activity (Caesar cipher) creates an awareness of knowledge gaps since for many of the students taking part in the experiment it would be the first time they need to decode something. They would need to start thinking about frequency analysis, which letters are the most likely to be repeated in a text and how to find patterns in a text in order to achieve the task.
For solving the second task, the same knowledge gaps will be raised but at the same time the student is expected to use what has just been explained in the Caesar cipher in order to solve a more challenging task since in this case the key is unknown as well as its length. This task is more demanding since the student is expected to transfer the Caesar cipher to a more complex scenario (the Vigenère cipher) and at the same time he needs to use the kasiski method to decrypt.


## Tips for your design
- Your learner does not need to solve your problem or learn from the problem solving activity directly! (PS-I is not Guided Discovery Learning)
- PS is to activate the prior knowledge and awareness of the knowledge gaps
- It is I following PS to explicitly facilitate the acquisition of the target knowledge
- Research suggests that "the problem is presented in the form of contrasting cases that differ in only one feature at a time" during the PS phase, or "if student solutions are discussed" during the I phase, to make PS-I effective.
- According to the literature, a good design of PS-I works better in acquiring conceptual knowledge and the ability to transfer.
