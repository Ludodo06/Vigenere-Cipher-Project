## What is the learning activity?

For example, **"Find the Most Reliable Pitching Machine"**

First, we give a code encoded with the ceasar cipher. We give them a ceasar cipher wheel, but do not explain what it is or how the code was encoded. 
We give them a hint, if they choose to use it, the first word of the plaintext is revealed.

Then we explain the ceasar cipher and how it works, adding what frequency analysis is and how it can be used to easily crack a substitution cipher. 

Then we explain what the Vigenère cipher is and how to use it, giving examples and a tool to encode and decode when the key is known. 

We then give them an encoded text without the key, and let the student try to decode it. ( We can give a tool that reorders the letters by dividing them in groups of the key length)

We then give them a hint, giving the size of the key. 

Finally once the exercise is done, we explain how to use the Kasiski examination method to break the Vigenère cipher.

## Tips for your design
- Your learner does not need to solve your problem or learn from the problem solving activity directly! (PS-I is not Guided Discovery Learning)
- PS is to activate the prior knowledge and awareness of the knowledge gaps
- It is I following PS to explicitly facilitate the acquisition of the target knowledge
- Research suggests that "the problem is presented in the form of contrasting cases that differ in only one feature at a time" during the PS phase, or "if student solutions are discussed" during the I phase, to make PS-I effective.
- According to the literature, a good design of PS-I works better in acquiring conceptual knowledge and the ability to transfer.
