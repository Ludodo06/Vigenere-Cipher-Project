# IMPORT DATAFRAME

```python
import pandas as pd

# Read the Excel file and store the data in a DataFrame
df = pd.read_excel('results.xlsx')

# Replace the comma with a dot and convert the column to float
df['Post test'] = df['Post test'].str.replace(',', '.').astype(float)
df['Learning gain'] = df['Learning gain'].str.replace(',', '.').astype(float)
df['Relative learning gain'] = df['Relative learning gain'].str.replace(',', '.').astype(float)

# Display the DataFrame
df
```

***
# IPS vs. PSI

```python
# Split the DataFrame into PSI and IPS DataFrames based on the condition column
psi_df = df[df['condition'] == 'PSI']
ips_df = df[df['condition'] == 'IPS']

# Display the PSI DataFrame
psi_df

# Display the IPS DataFrame
ips_df
```

```python
import matplotlib.pyplot as plt

# Compute means and std for "Pre test" and "Post test" for IPS dataframe
ips_pre_mean = ips_df['Pre test'].mean()
ips_pre_std = ips_df['Pre test'].std()
ips_post_mean = ips_df['Post test'].mean()
ips_post_std = ips_df['Post test'].std()


# Compute means and std for "Pre test" and "Post test" for PSI dataframe
psi_pre_mean = psi_df['Pre test'].mean()
psi_pre_std = psi_df['Pre test'].std()
psi_post_mean = psi_df['Post test'].mean()
psi_post_std = psi_df['Post test'].std()

# Plotting the means and std for "Pre test" and "Post test" for IPS and PSI dataframes
labels = ['I-PS', 'PS-I']
x = [0, 1]
pre_means = [ips_pre_mean, psi_pre_mean]
pre_stds = [ips_pre_std, psi_pre_std]
post_means = [ips_post_mean, psi_post_mean]
post_stds = [ips_post_std, psi_post_std]

# Shift the pre-means a bit to the left
x = [i-0.1 for i in x]

# Plotting means and std for "Pre test"
plt.errorbar(x, pre_means, yerr=pre_stds, fmt='o', label='Pre test')

# Shift the post-means a bit to the right
x = [i+0.2 for i in x]

# Plotting means and std for "Post test"
plt.errorbar(x, post_means, yerr=post_stds, fmt='o', label='Post test')

# Shift back the x-axis ticks and labels
x = [i-0.1 for i in x]

# Set the x-axis ticks and labels
plt.xticks(x, labels)

# Set the x-axis label
plt.xlabel('Group')
 
# Set the y-axis label
plt.ylabel('Test Score')

# Set the title of the plot
plt.title('Test Scores for I-PS and PS-I')

# Add a legend
plt.legend()

# Show the plot
plt.show()

```
